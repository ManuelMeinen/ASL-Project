```python
def get_limbs51(x):
  l = []
  for i in range(0,4):
    print(51*i,51*(i+1))
    v = int(x[51*i:51*(i+1)],2)
    l.append(v)
  print(len(x[51*4:256]))
  v = int(x[51*4:256],2)
  l.append(v)
  l.reverse()
  return l

def check(l, a):
  bstr = ""
  for i in [4,3,2,1,0]:  
    bstr = bstr + (str(bin(l[i]))[2:]).zfill(51)
  print(int(bstr,2))
  print(int(a,2))
```

```python
a = "111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111101101"
l_a = get_limbs51(a)
check(l_a,a)
print(l_a)

a_d = 13311224122668955738146552185873637781933051739291637575097427091064598392472
b_d = 11189996199370894582589062329635885928504840703623090769010532534865250839746
gmp_res = (a_d * b_d)%(2**255 - 19)
print(gmp_res)
print(str(bin(gmp_res))[2:])
l = get_limbs51((str(bin(gmp_res))[2:]).zfill(255))
check(l,str(bin(gmp_res))[2:].zfill(255))
print(l)
```
