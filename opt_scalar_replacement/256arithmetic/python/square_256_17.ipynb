```python
def gen_code(prod):
    code = ""
    for i in range(0, NO_LIMBS - 1):
        code += "_2a[%d] = a[%d] << 1;\n" % (i,i)
    code += "\n"
    
    i = 0
    for p in prod:
        code += gen_line(p[0],p[1],i)
        i += 1
    return code
```

```python
def gen_line(l_2a,l_a, i):
    l = "r[%d] = " %(i)
    for x in l_2a:
        l += "_2a[%d]*a[%d] + " % (x[0],x[1])
    
    for x in l_a:
        l += "a[%d]*a[%d] + " % (x[0],x[1])
    
    return l[:-3] + ";\n"
```
